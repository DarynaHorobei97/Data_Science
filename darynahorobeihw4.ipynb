```python
import numpy as np
import pandas
import math

print(np.__version__)
print(pandas.__version__)
'''print(math.__version__)'''

import random
import statistics as st

mean_list = []
sum_list = []
for n in range(1, 101):
    list_ = []
    for i in range(100):
        a = round(random.uniform(1, 100), 2)
        list_.append(a)
    print('List number', n, list_)
    mean_= round(st.mean(list_), 2)
    sum_= round(math.fsum(list_), 2)
    print('Mean value in the list is', mean_)
    print('Sum of the list elements is', sum_)
    print()
    mean_list.append(mean_)
    sum_list.append(sum_)
print('List with mean values of 100 lists is', mean_list)
print('List with sum values of 100 lists is', sum_list)
```
