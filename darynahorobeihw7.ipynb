```python
sql_query_matching = '''
SELECT Table2.Driver_name, Table2.Driver_category, Table1.Vehicle_type
FROM Table2
JOIN Table1 ON Table2.Table1_ID = Table1.ID
WHERE (Table2.Driver_category = 'A' AND Table1.Category = 'A') OR
      (Table2.Driver_category = 'B' AND Table1.Category = 'B') OR
      (Table2.Driver_category = 'C' AND Table1.Category = 'C');
'''

matching_df = pd.read_sql_query(sql_query_matching, connection)

matching_df.to_csv('matchDrivers.csv', index=False)
```

```python
sql_query_roads = '''
SELECT DISTINCT Table2.Driver_name, Table2.Driver_category, Table1.Vehicle_type
FROM Table2
JOIN Table1 ON Table2.Table1_ID = Table1.ID
JOIN Table3 ON Table1.Category = Table3.Category
WHERE (Table3.Permitted_roads = 'PO87' OR Table3.Permitted_roads = 'QWE7')
AND (Table2.Driver_category = Table1.Category)
AND (Table1.Category = 'A' OR Table1.Category = 'B' OR Table1.Category = 'C')
GROUP BY Table2.Driver_name, Table2.Driver_category, Table1.Vehicle_type
HAVING COUNT(DISTINCT Table3.Permitted_roads) = 2;
'''

roads_df = pd.read_sql_query(sql_query_roads, connection)

roads_df.to_csv('driversPermittedRoads.csv', index = False)
```

```python
sql_query_avg_speed = '''
SELECT Category, AVG(Max_speed) AS Average_Speed
FROM Table1
GROUP BY Category
'''

avg_speed_df = pd.read_sql_query(sql_query_avg_speed, connection)
avg_speed_df
```
